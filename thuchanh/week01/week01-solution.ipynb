```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('patient_heart_rate.csv')
df.head()
```

```python
df[['Firstname', 'Lastname']] = df['Name'].str.split(expand= True)
df = df.drop(columns = ['Name'])
df.head()
```

```python
weight = df['Weight']
for i in range(0, len(weight)):
    x = str(weight[i])
    if 'lbs' in x[-3:]:
        x = x[:-3]
        float_x = float(x)
        y = int(float_x/2.2)
        weight[i] = y
    if 'kgs' in x[-3:]:
        x = x[:-3:]
        float_x = float(x)
        weight[i] = x
```

```python
df.rename(columns={'Weight': 'Weight_kg'}, inplace=True)
df.head()
```

```python
df.dropna(how='all', inplace=True)
```

```python
df =df.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight_kg'])
```

```python
df['Age'].fillna(df['Age'].mode()[0], inplace=True)
```
